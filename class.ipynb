# Creating an instance
- Usually, when creating an instance of a type, we will want to provide some initial values for the internal data. To do this, define an init method

```python
class Coordinate(object): #overriding
    def __init__(self, x, y): #3 arguments self points the frame
        self.x = x
        self.y = y
```

- The "." operator is used to access an attribute of an object. So the init method above is defining two attributes for the new __Coordinate__ object: x and y.

- there are two instance variables for any instance of a coordinate (data attribute)

```python
c = Coordinate(3,4)
origin = Coordinate(0,0)
print c.x, origin.x
```

- One can define a str method for a class, which python will call when it needs a string to print. This method will be called with the object as the first argument and should return a str.

```python
class Coordinate(object): #overriding
    def __init__(self, x, y): #3 arguments self points the frame
        self.x = x
        self.y = y
    def __str__(self):
        return "<" + str(self.x) + "," + str(self.y) + ">"
```

```python
c = Coordinate(3,4)
```

```python
import math
class Coordinate(object): #overriding
    def __init__(self, x, y): #3 arguments self points the frame
        self.x = x
        self.y = y
    def __str__(self):
        return "<" + str(self.x) + "," + str(self.y) + ">"
    def distance(self,other):
        return math.sqrt(abs(self.x - other.x) + abs(self.y - other.y))
```

```python
c = Coordinate(3,4)
d = Coordinate(4,5)
```

```python
c.distance(d)
```
