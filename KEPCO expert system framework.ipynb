os 모듈을 임포트하고 엑셀 파일이 있는 곳으로 상대 디렉토리 지정. xlrd 라는 모듈 필요 (pip install xlrd), xlrd 임포트 한뒤 .open_workdbook  method로 파일 열기

```python
import pprint
class MyPrettyPrinter(pprint.PrettyPrinter): #한글폰트를 프린트 하는 class
	def format(self, _object, context, maxlevels, level):
		if isinstance(_object, unicode):
			return "'%s'" % _object.encode('utf8'), True, False
		elif isinstance(_object, str):
			_object = unicode(_object,'utf8')
			return "'%s'" % _object.encode('utf8'), True, False
		return pprint.PrettyPrinter.format(self, _object, context, maxlevels, level)
```

python 2.7 에서 unicode print를 가능하게 하는 class

```python
from xlrd import open_workbook
book = open_workbook('rulebase.xlsx')
sheet = book.sheet_by_index(0)

for j in range(sheet.nrows):
    rules = [[] for i in range(sheet.nrows)]

for j in range(sheet.nrows):    
    for i in range(sheet.ncols):
        rules[j].append(sheet.cell_value(j,i))
```

sheet_by_index(0) 은 엑셀 파일 중 가장 첫번째 index(0)을 현재 작업할 시트로 지정하는 method, .sheet_cell_value(row,column) 는 지정 한 row 와 column의 cell을 return하는 method

원래 아이디어는 엑셀 시트의 크기만한 array를 만들어서 엑셀의 for 문을 2번 돌면서 각 array의 칸에 대응하는 cell을 입력하려고 했으나, 현재 엑셀을 __한글__로 작성했기 때문에 array의 인덱스에 맞춰서 값을 선언해주는 법이 불가 -> 방법은 있으나 여러 추가 모듈을 더 설치해야 하거나 unicode를 글자 단위로 다시 쪼갠뒤 str값으로 변환해야하기 때문에 패스

unicode가 리스트로는 선언가능한 것에 착안, columns를 하나의 리스트로 보고 하나의 큰 리스트를 만든뒤 그 안에 엑셀 시트의 row의 갯수만큼 리스트를 다시 만듦(for j in(sheet.nrows)). 큰 리스트 안에 길이 5개 짜리 작은 리스트를 만들어서 cell의 값을 return 하여 받음

```python
MyPrettyPrinter().pprint(rules) #큰 리스트의 구조
```

```python
def infengine1(machine,status,rule):
    for i in range(len(rules)):
        if rule[i][0] == machine and rule[i][1] == status:
            return 'Cause: ' + rule[i][2], 'Do: ' + rule[i][3]
```

그후 일단 3가지 패러미터를 받는 inference engine1을 만듦. machine(진단하고자 하는 기계), status(증상), rule(지식기반데이터)를 받으며 rule안
에서 사용자가 입력한 machine,status의 값을 찾음. u' 유니코드로 입력하는 스트링, 3가지 조건을 테스트 해봄

```python
MyPrettyPrinter().pprint(infengine1(u'BFP',u'진동이 크다',rules))
```

```python
MyPrettyPrinter().pprint(infengine1(u'BFP',u'반구동측의 진동증가',rules))
```

```python
MyPrettyPrinter().pprint(infengine1(u'수직펌프',u'위험속도를 넘어서 운전하면 진동발생',rules))
```
