# Introduction to Fuzzy Logic

- Fuzzy logic is a form of many-valued logic which deals with reasoning that is approximate rather than fixed and exact
- Compared to traditional binary sets (where variables may take on true or false values), fuzzy logic variables may have a truth value that ranges in degree __between 0 and 1__
- Fuzzy logic has been extended to handle the concept of partial truth, where the truth value may range between completely true and completely false
- Before going deepr to Fuzzy Logic, the understanding of Fuzzy Set must be preceded 

#Before coming to the details, let me introduce Simplest Rule-Based System


- The most widely used conrol systems is the simplest rule-based system imaginable, a thermostatic temperature controller
- This rule-based system operates with two rules:

$${\huge\begin{array} (1) IF \text{ temperature is below set point} \quad THEN \text{ heat is on}
\\ 2) IF \text{ temperature is above set point} \quad THEN \text{ heat is off}
\end{array}}$$

- The success of this controller is due to the combination of the properties, that it is simple, robust and does not require a complex process model
- The model is:

$${\huge\begin{array} (\text{When the heat is on, the temperature rises slowly, and when the heat is off, the temperatures falls slowly} 
\end{array}}$$

The two IF-THEN clauses above can also be formally rewritten as

$${\huge\begin{array} ( IF \quad x \text{ is} \quad A_r \quad THEN \quad \mu \quad \text{is } B_r \text{  for } r = 1,2 \end{array}}$$

- In the thermostat example there is only one input variable (inguistic variable), the temperature *x*. In the general case, there are several input variables *x*<sub>1</sub>,...,*x*<sub>n</sub>, so, in addition to the logical connective IF-THEN, another logical connective is needed, AND.
- Then the IF-THEN clauses are

$${\huge\begin{array} ( IF \quad x \text{ is} \quad A_{r1} \text{ AND } A_{r2}... \text{AND} \quad x_n \text{ is } A_{rn} \quad THEN \quad \mu \quad \text{is } B_r \text{  for } r = 1,2 \end{array}}$$

- Here, *r = 1, ..., R* is the rule number, and *A*<sub>ri</sub> and *B*<sub>ri</sub> are words from natural language (linguistic terms), like "below set point", "on", "small", "large", "approximately 1.5", etc
- If the standard mathematical notation for IF-THEN and AND is used, the above rules can be  re-formulated as follows:

$${\huge\begin{array} ( A_{r1}(x_1)\wedge A_{r2}(x_2)... \wedge A_{rn}(x_n) \rightarrow B_r(\mu) \end{array}}$$

- Where *r = 1,...,R*. The set of rules is usually called a __rule base__. The left-hand side of a rule is called __premise__, the right hand side __conclusion__ and the rule itself __implication__

- The general idea is to represent the rule base in a computer, it has clear structure. A rule base consists of rules and each rule, in its turn, is obtained from properties expressed by linguistic variables and terms and using logical connectives
- In view of this structure, it is reasonable to represent the rule base by first representing the basic elemenets of the rule base, premises and conclusions, and then by extending this representation to the rule base as a whole

__Following steps in the methodology:__
1. Representation of the basic properties *A*<sub>ri</sub>(*x*<sub>i</sub>) and *B*<sub>r</sub>(&mu;)
2. Representation of the logical connectives
3. The representations of the basic properties and of the logical connectives is used to get the representations of all the rules
4. Combination of the representations of different rules into a representation of a rule base

- As a result of these four steps, one obtains an expert system that can give advise for a specialist, who has to make a decision
- From now on, this page will cover how these 4 steps are implemented

## Fuzzy Set

Fuzzy set theory is an extension of classical set theory. Under this aspect the term of the membership function is introduced. The operations with fuzzy sets are introduced and explained and illustrated using membership functions. Starting with elementary operators for fuzzy sets, more complicated operators for modelling of rule bases and reasoning processes are introduced


In the classicals set theory, a set can be represented by enumerating all its elements using

$${\huge\begin{equation}\ A = \{\ a_1,a_2,a_3,...,a_n \}\ \end{equation}}$$

If these elements *A* = {a<sub>1</sub>,a<sub>2</sub>,a<sub>3</sub>,...,a<sub>n</sub>} of *A* are together a subset of the universal base set *x*, the set *A* can be represented for all elements *x* &isin; *X* by its __characteristic function__

$${\huge\
 \mu_a(x)\ = \left\{ 
  \begin{array}{l l}
    1 & \quad \text{if } x \in A\\
    0 & \quad \text{otherwise}
  \end{array} \right.}$$

In classical set theory &mu;<sub>A</sub>(*x*) has only the values 0 ('False') and 1 ('True'), so two values of truth
Such sets are called __Crisp Sets__

- Non - Crisp sets are called __Fuzzy Sets__, for which also a characteristic funtion can be defined
- This function is a generalisations of characteristic funtion and called a __membership funtion__
- The membership of a fuzzy set is described by this membership function &mu;<sub>A</sub>(*x*) of A, which associate element *x<sub>0<sub>* &isin; *X* a grade of membership &mu;<sub>A</sub>(*x<sub>0</sub>*)
- In contrast to classical set theory a membership function of a fuzzy set can have in the normalised closed interval [0,1]

```python
#Plotting of Grade of truth of binary valued logic {0,1} and fuzzy logic {0,1}

import pylab as py
x = linspace(0,200,200)
binary = []
fuzzy = []
for i in x:
    if i < 100.0:
        binary.append(0.0)
    else:
        binary.append(1.0)

for j in x:
    if j < 80.0:
        fuzzy.append(0.0)
    elif j > 80.0 and j < 120.0:
        fuzzy.append(x[j]*1/40 - 2)
    if j > 120:
        fuzzy.append(1.0)
        
v = [0,200,0,1.2]

py.axis(v)
figure(1)
py.plot(x,binary)
py.plot(x,1-array(binary),'r')
xlabel('Height',fontsize =14)
title('Binary valued logic {0,1}',fontsize = 18)
text(40,1.05,'not tall')
text(140,1.05,'tall')
figure(2)
py.axis(v)
py.plot(x,fuzzy,'b')
py.plot(x,1-array(fuzzy),'r')
xlabel('Height',fontsize =14)
title('Fuzzy Logic [0,1]',fontsize = 18)
text(40,1.05,'not tall')
text(140,1.05,'tall')
py.show()
```

- Therefore, each membership function maps elements of a given universal base set x, which is itself a crisp set, into real numbers in [0,1]
- The notation for the membership function &mu;<sub>A</sub>(*x*) of a fuzzy set *A*

$${\huge\begin{equation}\ A:X \rightarrow [0,1] \end{equation}}$$

- Each Fuzzy Set is completely and uniquely defined by one particular membership function
- Consequently symbols of membership functions are also used as labels of the associated __Fuzzy Sets__
- The same notaion is used for __Crisp Sets__

```python
#trapezoidal function

def trapez(x,a,b,c,d):
    if float(x) < float(a) or float(x) > float(d):
        return 0.0
    if float(x) >= float(a) and float(x) <= float(b):
        return (x - a)/float(b -a)
    if float(x) >= float(b) and float(x) <= float(c):
        return 1.0
    if float(x) >= float(c) and float(x) <= float(d):
        return (d - x)/float(d - c)

#test
v = [0,200,0,1.1]
py.axis(v)
py.hlines(1.0,0,200, linewidth = 3, linestyle = 'dotted', color = 'k')
x = linspace(0,200,200)
y = [trapez(x[i-1],10,20,180,190) for i in x]
py.plot(x,y)
title('Trapezoidal Function', fontsize = 16)
show()
```

```python
x = linspace(0,200,200)
for j in x:
    if j < 100:
        crisp = [trapez(x[i-1],20,20,80,80) for i in x]
    else:
        fuzzy = [trapez(x[i-1],120,130,170,180) for i in x]

v = [0,200,0,1.2]
py.axis(v)        
py.plot(x[0:100],crisp[0:100])
py.plot(x[100:200],fuzzy[100:200],'r')
title('Membership functions of a crisp set C and a fuzzy set F',fontsize = 20)
text(45,1.05,r'$\mu_c(x)$')
text(145,1.05,r'$\mu_F(x)$')
py.show()
```

- The base set *X* is introduced first above as a universal set
- In practical applications, physical or similar quantites are considered that are defined in some interval
- When Such quantities are described by sets, a base sets can be generalised seamless to a crisp base set *X* that exists in a defined interval
- Base sets are not always crisp sets
- Another generalisation is that the base set is itself a fuzzy set

## Membership Functions

- The membership function &mu;<sub>A</sub>(*x*) describes the membership of the elemetns *x* of the base set *X* in the Fuzzy Set *A*
- For &mu;<sub>A</sub>(*x*), a large class of functions can be taken
- Reasonable functions are often piecewise linear function, such as triangular or trapezoidal functions
- The grade of membership &mu;<sub>A</sub>(*x<sub>0</sub>*) of a membership function &mu;<sub>A</sub>(*x*) describes the special element *x* = *x*<sub>0</sub>, to which grade it belongs to the Fuzzy Set *A*
- This value is inthe unit interval [0,1] and *x*<sub>0</sub> can simultanesously belong to another fuzzy set *B*, such that &mu;<sub>B</sub>(*x*<sub>0</sub>) characterises the grade of membership of x<sub>0</sub> to *B*

```python
import pylab as py

x = linspace(0,200,200)
fuzzy1 = [trapez(x[i-1],10,20,80,150) for i in x]
fuzzy2 = []
for j in x:
    if j < 80.0:
        fuzzy2.append(0.0)
    elif j > 80.0 and j < 120.0:
        fuzzy2.append(x[j]*1/40 - 2)
    if j > 120:
        fuzzy2.append(1.0)

v = [0,200,0,1.2]
py.axis(v)
figure(1)
py.plot(x,fuzzy1,'r')
py.plot(x,fuzzy2,'b')
title('Membership grades of x0 in the sets A and B',fontsize = 18)
text(67,1.05,r'$\mu_A$')
text(150,1.05,r'$\mu_B$')
py.show()
```

- In the following, a set of important propoerties and characteristics of Fuzzy Sets will be described
- Having two Fuzzy Sets, *A* and *B* based on *X*, then both are __equal__ if their membership functions are equal, i.e.

$${\huge\begin{equation}\ A = B \leftrightarrow \mu_A(x)\ = \mu_B(x)\, x \in\ X \end{equation}}$$

The __universal set *U*__ is defined as

$${\huge\begin{equation}\mu_U(x)\ = 1,\quad x \in\ X \end{equation}}$$

The __height__ of a Fuzzy Set *A* is the largest membership grade obtained by any element in that set, i.e.(In this context, sup is considered as max)

$${\huge\begin{equation}\ \text{hgt(A)}\ = \underset{x\in{X}}{\text{sup}}\mu_A(x)\ \end{equation}}$$

A Fuzzy Set *A* is called __normal__ when hgt(A) = 1, and it is __subnormal__ when hgt(A) < 1

The __support__ of a Fuzzy Set *A* is the crisp set that contains all the elements of *X* that have nonzero membership grades in *A*, i.e.

$${\huge\begin{equation}\ \text{supp(A)}\ = \{x \in\ X|\mu_A(x)\ > 0 \}\ \ \end{equation}}$$

The __core__ of a normal Fuzzy Set *A* is the crisp set that contains all the elements of *X* that have the membership grades of one in *A*, i.e.

$${\huge\begin{equation}\ \text{core(A)}\ = \{x \in\ X|\mu_A(x)\ = 1 \}\ \ \end{equation}}$$

The __boundary__ is the crisp set that contains all the elements of *x* that have the membership grades of 0 <&mu;<sub>A</sub>(x)< 1 in *A*, i.e.

$${\huge\begin{equation}\ \text{bnd(A)}\ = \{x \in\ X|0 < \mu_A(x)\ < 1 \}\ \ \end{equation}}$$

Having two Fuzzy Sets *A* and *B* based on *X*, then both are __similar__ if

$${\huge\begin{equation}\ \text{core(A)}\ = \text{core(B)}\quad and \quad \text{supp(A)}\ = \text{supp(B)}\ \end{equation}}$$

If the support of a normal Fuzzy Set consists of a single element *x*<sub>0</sub> of *X*, which has the property

$${\huge\begin{equation}\ \text{core(A)}\ = \text{core(B)}=\{x_0\}\\ \end{equation}}$$

This set is called a __singletone__

- The type of representation of the membership function depends on the base set. If this set consists of many values, or is the base set a continuum, then a __parametric representation__ is appropriate. For that functions are used that can be adapted by changing the parameters
- Piecewise linear membership functions are preffered, because of their simplicity and efficiency with respect to computability
- Mostly these are trapezoidal or triangular functions, which are defined by four and three parameters, respectiveley

$${\huge\
 \mu(x,a,b,c,d)\ = \left\{ 
  \begin{array}\
    0,\quad x < a, x> d\\
    \frac{x-a}{b-a},\quad a \leq x \leq b\\
    1,\quad b \leq x \leq c\\
    \frac{d-x}{d-c},\quad c \leq x \leq d
  \end{array} \right.}$$

- Migrates for case *b* = *c* into a triangular function
- For some applications the modeling requires continuously differentiable curves and therefore smooth transition, which the trapezoids do not have

## Elementary operators for Fuzzy Sets

- The basic connective operations in classical set theory are those of intersection, union and complement
- These operations on characteristic functions can be generalised to Fuzzy Sets in more than one way
- However, one particular generalisation, which results in operations that are usually referred to us as standard Fuzzy Set operations, has a special significance in Fuzzy Set theory

The __Fuzzy Intersection__ operator &cap; (Fuzzy AND connective appliied to two Fuzzy Sets *A* and *B* wit the membership functions &mu<sub>A</sub>(*x*) and &mu<sub>B</sub>(*x*) is

$${\huge\begin{equation}\mu_{A \cap B}(x)\ = \text{min} \{\ \mu_A(x), \mu_B(x) \}\ , x \in X \
, \\ \mu_{A \cap B}(x)\ = \mu_A(x)\mu_B(x)
\end{equation}}$$

```python
# Fuzzy AND
x = linspace(1,200,200)
fuzzy1 = [trapez(x[i - 1],50,80,80,110) for i in x]
fuzzy2 = [trapez(x[i - 1],90,120,120,150) for i in x]
AND1 = array([fuzzy1,fuzzy2])
AND1 = [min(AND1[0,i],AND1[1,i]) for i in range(200)]
AND2 = array(fuzzy1)*array(fuzzy2)
plot(x,fuzzy1,label = 'Fuzzy1')
plot(x,fuzzy2,'r' , label = 'Fuzzy2')
plot(x,AND1,'k',linewidth = 3, label = 'AND1')
plot(x,AND2,linewidth = 3, label = 'AND2')
title('Fuzzy AND', fontsize = 14)
legend()
show()
```

The __Fuzzy Union__ operator &cup; (Fuzzy OR connective) applied to two Fuzzy Sets *A* and *B* wit the membership functions &mu;<sub>A</sub>(x) and &mu;<sub>B</sub>(x) is

$${\huge\begin{equation}\mu_{A \cup B}(x)\ = \text{max} \{\ \mu_A(x), \mu_B(x) \}\ , x \in X,
\\ \mu_{A \cup B}(x)\ = \mu_A(x) + \mu_B(x) - \mu_A(x) \mu_B(x)
\end{equation}}$$

```python
# Fuzzy OR
x = linspace(1,200,200)
fuzzy1 = [trapez(x[i - 1],50,80,80,110) for i in x]
fuzzy2 = [trapez(x[i - 1],90,120,120,150) for i in x]
OR1 = array([fuzzy1,fuzzy2])
OR1 = [max(OR1[0,i],OR1[1,i]) for i in range(200)]
OR2 = array(fuzzy1) + array(fuzzy2) - array(fuzzy1)*array(fuzzy2)
plot(x,fuzzy1,label = 'Fuzzy1')
plot(x,fuzzy2,'r' , label = 'Fuzzy2')
plot(x,OR1,'k',linewidth = 3, label = 'OR1')
plot(x,OR2,linewidth = 3, label = 'OR2')
title('Fuzzy OR', fontsize = 14)
legend()
show()
```

The __Fuzzy Complement__ (Fuzzy NOT operation) applied to the Fuzzy Set *A* wit the membership function &mu;<sub>A</sub>(x) is

$${\huge\begin{equation}\mu_\bar A\ = 1 - \mu_A(x) , x \in X
\end{equation}}$$

```python
# Fuzzy NOT
x = linspace(1,200,200)
fuzzy = [trapez(x[i - 1],30,80,80,120) for i in x]
nfuzzy = 1 - array(fuzzy)
v = [0,200,0,1.1]
axis(v)
hlines(1.0,0,200, linewidth = 3, linestyle = 'dotted')
plot(x,fuzzy,label = 'Fuzzy')
plot(x,nfuzzy,'r' , label = 'Fuzzy NOT')
title('Fuzzy OR', fontsize = 14)
text(20, 0.2, r'$\mu_a(x)$',fontsize = 14)
text(115, 0.7, r'$ 1 -\mu_a(x)$',fontsize = 14)
legend()
show()
```

- These operations perform precisely as the corresponding operations for __Crisp Sets when the range of membership grades is restricted to the set {0,1}__
- That is, the standard Fuzzy Operations are generalisations of the corresponding classical set operations
- However, they are not the only possible generalisation. As shown above, the fuzzy intersection, __union and complement are not unique operations__, contrary to their crisp counterparts. Different functions may be appropriate to represent these operations in different contexts. __The capability to determine appropriate membership functions and meaningful fuzzy operations in the context of each particular application is crucial__ for making fuzzy set theory practically useful

## Fuzzy Relations

- Relations are explained by a simple example discrete fuzzy sets.
- Let's describe the relationship between the __genre of a game *x*__ and __the grade of difficulties of game *y*__ and characterise the linguistic variable color by a crisp set *X* with three linguistic terms as

$${\huge\begin{equation} X = \{\ Action, RPG, Simulation \}\
\end{equation}}$$

and the similary the grade of difficulties of game as

$${\huge\begin{equation} Y = \{\ Easy, Normal, Hard \}\
\end{equation}}$$

One knows that a crisp formulation of a relation *X* -> *Y* between the two crisp sets would look like this in tabular form:

|   |Easy   |Normal   |Hard   |
|---|---|---|---|
|Action   |1   |0   |0   |
|RPG   |0   |1   |0   |
|Simulation   |0   |0   |1   |

The zeros and ones describe the grade of membership to this relation. This relation is now a new kind of crisp set built from the two crisp sets *X* and *Y*. This new set is now called *R* can be expressed also by the rules:

$${\huge\begin{array} (1) IF \text{ the genre is Action} \quad THEN \text{ the game is easy}
\\ 2) IF \text{ the genre is RPG} \quad THEN \text{ the game is normal}
\\ 3) IF \text{ the genre is Hard} \quad THEN \text{ the game is hard}
\end{array}}$$

This crisp relation *R* represents the presence or absence of association, interaction or interconnection between the elements of these two sets. This can be generalised to allow for various degrees of strength of association or interaction between elements. Degree of association can be represented by membership grades in a fuzzy relation in the same way as degrees of the set membership are represented in a fuzzy set. Applying this to the game example, the table can be modified to:

|   |Easy   |Normal   |Hard   |
|---|---|---|---|
|Action   |1   |0.5   |0   |
|RPG   |0.3   |1   |0.4   |
|Simulation   |0   |0.2   |1   |

Where there are now real numbers in [0,1]. This table represents a fuzzy relation and models the connectives in a fuzzy rule base. It is a two-dimensional fuzzy set and __the question now is, how can this set be determined from its elements.__

- In order to do this, the elements are generalised
- In the above example, the linguistic terms where treated as crisp terms
    - For example, when one represents the colors on color spectrum scale, the colors would be describe by their spectral distribution curves that can be interpreted as membership functions and then a particular color is a fuzzy term
- Treating also the grades of maturity as fuzzy terms, the above relation is a two-dimensional fuzzy set over two fuzzy sets
    - For example, taking from the game example the relation between the linguistic terms __Simulation__ and __Hard__, and represent them by the membership functions property __Simulation__ and __Hard__

- Which descrbies in a linguistic way a union of three rules. For the complete rule base 'R' one can combine the relations formed for each individual rule with a fuzzy union operator, which is the fuzzy OR
- Now, __Combination of the representations of different rules into a representation of a rule base__ can be specified by taking the rule base and applying the union operator by writing the rule base with max/min operators as follows:

$${\huge\begin{array} ( \mu_R(x_1, x_2, x_3...,x_n,u) = \underset{r}{\text{max}} \{{\text{min}} \{\mu_{P_r}(x_1,x_2,...,x_n), \mu_{B_r} \} \} \end{array}}$$

where &mu;<sub>P<sub>r</sub></sub>(*x*<sub>1</sub>, *x*<sub>2</sub>,...,*x*<sub>n</sub>) is the premise of the <sub>r</sub>th rule. This representation is the standard max/min representation of a rule base that will be later used. Instead of the max/min representation a so called max-prod representation is also usual, where the algebraic product

$${\huge\begin{array} ( \mu_R(x_1, x_2, x_3...,x_n,u) = \underset{r}{\text{max}} \{ \mu{P_r}(x_1,x_2,...,x_n) \mu_{B_r} \}  \end{array}}$$

## Fuzzy Composition

- Based on facts for *x*<sub>i</sub> and on the rule base a reasoning procedure make a decision. xplaining reasoning by the fruit example, it is assumed that one has a crisp fact: __RPG game__. The decision from the rule base is obvious: __the difficulty is Normal__, and this is similar for the other facts: __Action and Simulation__

- But if one has a fact like: the genre of game is __Action-RPG__, one does not know how to determine which rule fires the decision and what the decision is. In the following, a new fuzzy operation type is introduced, that allows to operate with a given fact and a fuzzy relation to produce an output that represents the decision in fuzzy way

- This operation is called __Fuzzy Reasoning__, which is a special case of the more general operation called __Fuzzy Composition__

Two relations of the form given

$${\huge\begin{equation} R: X \times Y \rightarrow [0,1]
\\ S: Y \times Z \rightarrow [0,1]
\end{equation}}$$

can be composed to one relation

$${\huge\begin{equation} T: X \times Z \rightarrow [0,1]
\end{equation}}$$

- This process is known as composition and, using the max and min operators for union and intersection
- One can express the composition operator *T* = *X*&sdot;*Z* -> [0,1] by the corresponding membership functions as follows:

$${\huge\begin{array} ( \mu_R(x,z) = \underset{y \in Y}{\text{max}} \{ \text{min} \{\mu_R(x,y), \mu_S(y,z) \} \}  \end{array}}$$

When one takes the above game example again with the genre-difficulty relation *R*

|*R*   |Easy   |Normal   |Hard   |
|---|---|---|---|
|Action   |1   |0.5   |0   |
|RPG   |0.3   |1   |0.4   |
|Simulation   |0   |0.2   |1   |

and define for *S* a difficulty-fun relation

|*S*   |Very Fun   |Not Bad   |Sucks   |
|---|---|---|---|
|Easy   |1   |0.2   |0   |
|Normal   |0.7   |1   |0.3   |
|Hard   |0   |0.7   |1   |

then by applying above equation to the elements of these two tables, the following is obtained

|*T* = *R*&sdot;*S*   |Very Fun   |Not Bad   |Sucks   |
|---|---|---|---|
|Action   |1   |0.5   |0.3   |
|RPG   |0.7   |1   |0.4   |

$${\huge\begin{array} ( \mu_R(x,z) = \underset{y \in Y}{\text{max}} \{ \text{min} \{\mu_R(x,y), \mu_S(y,z) \} \}  \end{array}}$$

is obtained. The result is obvious from the first rule of the rule base *R*. When a different color is taken than included in the rule base entries, say __Action-RPG__ as

$${\huge\begin{array} ( C' = \{0 \quad 0.5 \quad 0.5 \} \end{array}}$$

then there is no problem to obtain the value for the difficulty

$${\huge\begin{array} ( T' = \{0.3 \quad 0.5 \quad 0.5 \} \end{array}}$$

In the same manner as relations can be composed, the one-dimensional facts can be composed with the rule base to realise the reasoning operation. This can now be precisely re-formulated for the general case of a rule base according to the Equation

$${\huge\begin{array} ( A_{r1}(x_1)\wedge A_{r2}(x_2)... \wedge A_{rn}(x_n) \rightarrow B_r(\mu) \end{array}}$$

If for the rule base

$${\huge\begin{array} ( R:X \times Y \rightarrow [0,1]\end{array}}$$

its membership function is described by equation

$${\huge\begin{equation}\mu_{A \cap B}(x)\ = \text{min} \{\ \mu_A(x), \mu_B(x) \}\ , x \in X \
, \\ \mu_{A \cap B}(x)\ = \mu_A(x)\mu_B(x)
\end{equation}}$$

if there is a fact described by the fuzzy set

$${\huge\begin{array} ( A':X \rightarrow [0,1]\end{array}}$$

and its membership function &mu;<sub>A</sub>(x), the result

$${\huge\begin{array} ( B' = A' \circ R: Y \rightarrow [0,1]\end{array}}$$

of the Fuzzy Reasoning is represented by the membership function

## Fuzzy Inference Machine

The core section of a fuzzy is that part, which combines the facts obtained from the fuzzification with the rule base and conducts the fuzzy reasoning process. This is called a __Fuzzy Inference Machine__. 

In the following, forsimplicity it is assumed that there is only one input *x*<sub>1</sub> = *x* and the rule base is described with max/min operators

$${\huge\begin{array} ( \mu_R(x,u) = \underset{r}{\text{max}} \{{\text{min}} \{\mu_{P_r}(x), \mu_{B_r} \} \} \end{array}}$$

then inserting this equation to 

$${\huge\begin{array} ( \mu_B'(y) = \underset{x \in X}{\text{max}} \{ \text{min} \{ \mu_A'(x), \mu_R(x,y) \} \} \end{array}}$$

yields

$${\huge\begin{array} ( \mu_B'(y) = \underset{x \in X}{\text{max}} \{ \text{min} \{ \mu_A'(x), \underset{r}{\text{max}} \{{\text{min}} \{\mu_{P_r}(x), \mu_{B_r} \} \} \} \end{array}}$$

The __max__ operations can be reordered such that only the relevant operands are on the right-hand side. Then

$${\huge\begin{array} ( \mu_B'(y) = \underset{r}{\text{max}} \{ \text{min} \{ \underbrace{ \underset{x \in X}{ \text{max}} \{ \text{min} \{ \mu_{A'}(x), \mu_{P_r}(x) \} \} , \mu_{B_r}(u) \} \}} _{H_r}\end{array}}$$

$${\huge\begin{array} ( \mu_B'(y) = \underset{r}{\text{max}} \{ \text{min} \{H_r, \mu_{B_r},(u) \} \} \end{array}}$$

is obtained for the reasoning proces. The inner term *H*<sub>r</sub>, which combines the fact with the premise, is a constant and is called __degree of relevance__ of the rule *r*. It characterises the relevance of the fired rule *r* and can be treated as a de-normalized universal fuzzy set.

## Defuzzification

As a result of applying the previous steps, one obtains a fuzzy set &mu;<sub>B'</sub>(u) from the reasoning process that describes, for each possible value *u*, how reasonable it is to use this particular value. In other words, for every possible value *u*, one gets a grade of membership that describes to what extent this value *u* is reasonable to use. Using a fuzzy system as a controller, one wants to transform this fuzzy information into a single value *u'* that will actually be applied. This transformation from a fuzzy set to a crisp number is called __defuzzification__. It is not a unique operation as different approaches are possible.

__Center of gravity method (COG)__

AThis approach has its origin in the idea to select a value *u'* that, on average, would lead to the smallest error in the sense of a criterion. If *u'* is chosen, and the best value is *u*, then the error is *u'*-*u*. Thus to determiine *u'* the least squares method can be used. As weights for each square, one can take the grade of membership *u*<sub>B'</sub>(u) with which *u* is areasonable value. As a result one has to find

$${\huge\begin{array} ( u' = \underset{u'}{\text{argmin}} \int_u u_{B'}(u)(u'-u)^2 \mathrm{d}u  \end{array}}$$

Differentiating with respect to the unknown *u'* and equating the derivative to zero, the formula

$${\huge\begin{array} ( u' = \frac{\int_u u_{B'}(u) \mathrm{d}u}{\int_u u_{B'}(u)u_{B'}(u)}  \end{array}}$$

__Center of singleton method (COS)__

The defuzzification can be strongly simplified if the membership functions &mu;<sub>B<sub>r</sub></sub>(u) of the conclusions are singly defuzzified for each rule such that each function is reduced to a singleton that has the position *u'*, of the individual membership function's center of gravity. The center of singletons is calculated by using the degree of relevance as follows

$${\huge\begin{array} ( u' = \frac{\sum_R u'_r {H_r}}{\sum_R H_r}  \end{array}}$$

The simplification consists in that the singletons can be determined already during the design of the fuzzy system and that the membership function *u*<sub>B'</sub>(u) with its complicated geometry is no longer needed. The defuzzification using this formula is an approximation of the defuzzification

__The Takagi-Sugeno Fuzzy System__

In the rule bases described hitherto with the IF-THEN rules of this chapter, fuzzy sets both in the premises and in the conclusions are used. This kind of inference is called Mamdani inference. A modified inference scheme, developed by Takagi and Sugeno, represents the conlusions by functions. A rule of this form will be

$${\huge\begin{array} ( \text{IF } x_1 \text{ is } A_{r1} \text{ AND } x_2 \text{ is } A_{r2}... \text{ AND } x_n \text{ is } A_{rn} \text{ THEN } u = f_r(x_1,x_2,...,x_n)\end{array}}$$

The structure of the premises are the same as for the Mamdani inference. However, in the conclusion all linguistic terms *B*<sub>r</sub> are substitued by the functions *f*<sub>r</sub>, and therefore it is not necessary to define a priori linguistic terms *B*<sub>r</sub>(u) for the conclusions. The function *f*<sub>r</sub> represents a direct mapping form the input space *X*<sub>1</sub>&times;*X*<sub>2</sub>&times;...&times;*X*<sub>n</sub> with the input values *x*<sub>1</sub>,*x*<sub>2</sub>,*x*<sub>3</sub>,...,*x*<sub>n</sub> to the output space *u*.

The connective operation in a rule is in this case performed via the degree of relevance *H*<sub>r</sub> of the premise of the rule *R*, and the function *f*<sub>r</sub> in the conclusion. The finaloutput is determined as a weight mean value over all *R* rules according to

$${\huge\begin{array} ( u' = \frac{\sum_R {H_r}f_r(x_1,x_2,...,x_n)}{\sum_R H_r}  \end{array}}$$

The effort of performing a defuzzification is saved, as the crisp value *u'* is directly determined by the inference operation and this makes this method attractive

The Takagi-Sugeno fuzzy system builds an overall combination of functions *f*<sub>r</sub>, which are valid in some range. If the membership functions of the fuzzy sets in the premises are overlapping, the transition between the functions is always continuous. For the special case of linear functions

$${\huge\begin{array} ( f_r(x_1,x_2,...,x_n) = \sum_{v=1}^n {C_{rv}}x_v  \end{array}}$$

the coefficients *c*<sub>rv</sub> can be determined by some identification procedure
