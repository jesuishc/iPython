# 규칙기반 전문가 시스템 (Rule-based expert systems)

## 1. 지식이란 무엇인가?

기계가 지능을 요하는 문제를 풀려면 누군가는 그 해답을 알고 있어야함, 즉 특정분야에 대한 __노하우(Know-how)__를 누군가는 가지고 있어야함

__지식이란?__

- 지식은 어떤 주제나 분야에 대해 이론적, 혹은 실제로 이해하는 것, 혹은 알려진 사실들의 모음
- 지식을 소유한 사람을 __전문가(expert)__라고 함

__전문가란?__

- 특정 분야에 해박한 지식과 풍부한 경험을 쌓은 사람
- 그 분야는 제한적

__전문가의 사고방식__

- 대부분의 전문가는 자신의 지식을 문제 풀이에 관한 __규칙 형식__으로 표현
- 예제: 신호등을 보고 길을 건너야하는 외국인에게 도움을 주는 전문가적 사고 방식

- 이 IF-THEN 형식으로 표현한 문장을 __생성규칙__ 또는 줄여서 __규칙__이라고 함
- AI에서 '규칙'은 지식을 표현하는 가장 일반적인 방법으로 ,IF 부분에는 주어진 정보를, THEN 부분에는 기술한 어떤 행동으로 연결 하는 구조

## 2. 지식 표현 기법으로써의 규칙

규칙은 두 부분으로 구성되며 전건(antecedent, 전제(promise) 또는 조건(condition)) 이라고 하는 __IF 부분__과 후건(consequent, 결론(conclusion) 또는 행동(action)) 이라고 하는 __THEN 부분__이 있음

일반적으로, 하나의 규칙에는 __AND__나 __OR__이 쓰일 수 있음. 그러나 같은 규칙 안에서는 논리곱과 논리합을 섞어쓰지 말 것

규칙의 후건에도 여러 절이 있을 수 있음

규칙의 전건에서는 __객체__(object, 언어 객체(linguistic object))에 __값(value)__이 들어감. 앞의 길건너기 예에서는 언어 객체인 '신호등' 값에는 녹색이나 빨간색이 들어갈 수 있다. __연산자(operator)__는 객체를 판별하고 대입한다

```python
def drawmoney(age,money):
    if age < 18 and money > 1000:
        return 'Need Parents\' signature'
    elif age > 18:
        return 'Take the money' + money
```

```python
def cal_med_insurancefee(income):
    if income > 16238:
        return income*1.5/100
    else:
        return 'No need to pay insurance fee'
```

규칙은 관계, 추천, 지시, 전략, 휴리스틱을 표현할 수 있다

- __사용자 인터페이스(user interface)__는 문제의 답을 찾고 싶어 하는 사용자와 전문가 시스템 간의 통신 수단

## 4. 순방향 연결(Forward Chaining) 과 역방향 연결(Backward Chaining) 추론 기법

규칙기반 전문가 시스템에서, 분야 지식은 IF-THEN 생성 규칙의 집합이며 데이터는 현재 상황에 대한 지식의 집합으로 표현됨. 추론 엔진은 기반지식에 저장된 각각의 규칙을 데이터베이스에 있는 사실과 비교한 뒤 IF(조건) 부분이 사실과 일치하면 그 규칙은 점화되고, THEN(취해야 할 행동) 부분을 수행한다

```python
Image(filename ='rulebase3.jpg')
```

규칙의 IF 부분과 사실과의 일치는 __추론 사슬(inference chains)__을 생함, 추론 사슬은 전문가 시스템이 결론에 이르기 위해 규칙을 어떻게 적용했는지를 나타냄

__Example__

처음 데이터 베이스에는 사실 A,B,C,D,E를 포함하고 있으며, 기반지식에는 다음 세 개의 규칙만 포함된다고 가정

- 전문가 시스템이 사실 Z를 추론하기 위해 
    - 규칙 3을 사용하여 A 부터 X를 도출
    - 규칙 2를 사용하여 X와 B, E를 사용
    - 규칙 1을 통하여 Z를 도출

```python
Image(filename ='rulebase4.jpg')
```

규칙을 수행하는 주된 방법은 두 가지가 있는데, 하나는 __순방향 연결(forward chaining)__, 다른 하나는 __역방향 연결(backward chaining)__ 이다

__순방향 연결__

좀 더 자세한 설명을 위해 규칙을 다시 정의 (->는 IF와 THEN 부분을 나타냄)

```python
Image(filename ='rulebase5.jpg')
```

- 순방향 연결은 __데이터 지향(data-driven)__ 추론으로, 알려진 데이터에서 추론을 시작하여 순방향으로 진행해나감
- 한 번에 가장 좋은 규칙 하나만 실행되면 규칙이 점화되면 그 규칙은 데이터베이스에 새로운 사실을 추가함
- 어떠한 규칙이라도 한 번만 수행되고 일치-점화 싸이클은 더 이상 점화할 수 있는 규칙이 없으면 중단됨

- 순방향 연결은 정해진 목표와 관련 없는 많은 규칙들이 수행될 수 있음
- 이 경우엔 역방향 추론이 좀 더 적합함

__역방향 연결__

- 역방향 연결은 __목표 지향(goal-driven)__ 추론으로, 전문가 시스템이 목표(가정해)를 정하고, 추론 엔진이 이를 증명하기 위한 증거 찾기를 시도하는방법
- 먼저 원하는 해가 있는 규칙을 찾기 위해 기반 지식을 탐색함 (규칙은 __THEN(취해야 할 행동)__ 부분에 목표가 반드시 있어야 함)
- 목표로 한 규칙을 발견하고 __IF(조건)__ 부분이 데이터베이스에 있는 데이터와 일치한다면, 규칙은 점화되고 목표는 증명됨
- 만일 일치하지 않는 경우가 발생하더라도 작업 중이던 규칙은 제쳐두고 (__스택(stack)__에 놓아둠) 새로운 목표와 하위 목표를 설정함
- 추론 엔진은 현재의 하위 목표를 증명하기 위해 기반지식에서 더 이상 규칙을 발견할 수 없을 때까지 규칙을 스택에 쌓아올림

__Example__

```python
Image(filename ='rulebase6.jpg')
```

1 단계에서, 추론 엔진은 사실 Z를 추론하기 위해 기반지식을 탐색해서 THEN 부분에 설정을 목표가 있는 규칙을 찾음. 추론 엔진은 규칙 1:Y & D --> Z를 찾아서 스택에 넣음

2 단계에서, 추론 엔진은 하위 목표와 사실 Y를 설정하고 검증하기 위해 데이터 베이스를 검색하고 사실 Y가 없는 것을 확인함. THEN 부분에 사실 Y가 있는 규칙을 찾기 위해 기반 지식을 다시 검색하고 규칙 2: X & B & E --> Y를 찾고 스택에 넣음

3 단계에서, 추론 엔진은 새로운 하위 목표로 사실 X를 설정한 뒤 규칙 3:A --> X 를 찾아내고 스택에 넣음

4 단계에서, 추론 엔진은 데이터 베이스에서 사실 A를 찾고 규칙 3:A --> X가 점화되며 새로운 사실 X를 추론함

5 단계에서, 추론 엔진은 하위 목표인 사실 Y로 돌아가 규칙 2:X & B & E --> Y를 찾고 점화시킨 뒤 새로운 사실 Y를 추가함

6 단계에서, 시스템의 원래 목표 사실 Z를 추론하기 위해 규칙 1:Y & D --> Z를 찾고 데이터베이스에 있는 모든을 기반으로 규칙 1을 수행함

__순방향 VS 역방향__

- 정보를 수집한 뒤 그로부터 결론 도출: 순방향
- 가정해에서 시작한 뒤 이를 증명: 역방향

## 5. Python을 사용한 간단한 Expert System 구축

```python
import pprint
class MyPrettyPrinter(pprint.PrettyPrinter): #한글폰트를 프린트 하는 class
	def format(self, _object, context, maxlevels, level):
		if isinstance(_object, unicode):
			return "'%s'" % _object.encode('utf8'), True, False
		elif isinstance(_object, str):
			_object = unicode(_object,'utf8')
			return "'%s'" % _object.encode('utf8'), True, False
		return pprint.PrettyPrinter.format(self, _object, context, maxlevels, level)
```

```python
# * coding:utf-8 *
class createRule:
    
        """사용자가 룰을 추가할 때 사용하는 클래스.  IF (기종) is (발생현상) -> THEN (원인) 으로 구성되어 있음.
        사용자는 'IF', 'THEN" 을 입력함"""
        
        def __init__(self, IF, THEN):
            self.IF = IF
            self.THEN = THEN
            
        def __setitem__(self,IF,THEN):
            self.THEN[IF] = THEN
            
        def addDB(rule,rulelist): #DB에 룰을 추가하는 method, rulelist는 dict() 의 타입
            rulelist[rule.IF] = rule.THEN
            return rulelist
        
def inferNov(IF,DB): #DB에서 IF 조건을 검색
    return DB[IF]
            
```

```python
rule = [createRule(u'BFP 진동이 크다',u'펌프 유체력이 작용해서 베어링 유막 동특성이 변하여 위험속도가 저하'),
            createRule(u'BFP 반구동측의 진동증가',u'환상씰, 유체불안전화력'),
            createRule(u'BFP 년수경과 후 진동이 커짐', u'임펠러 유체력에 의한 저주파 진동'),
            createRule(u"BFP 위험속도 통과 후 진동 발생", u'축계의 내부마찰'),
            createRule(u'LNG용 원심 압추기 질소가스씰 이상으로 정지후, 재기동했을때 정격회전수의 1/2 이하의 회전수에서 진동발생', u'씰의 고장으로 외기를 흡입하여 국부적으로 130도의 온도차가 발생 로터가 열적 굽힙을 일으킴'),
            createRule(u'가변속 팬 정격회전수 부근에서 진동이 크다', u'디퓨저의 선회실속으로 가진력 발생'),
            createRule(u'다단 터빈펌프 저유량 영역에서 저주파수 진동발생', u'디퓨저의 선회실속으로 가진력 발생'),
            createRule(u'다단 터빈펌프 토출유량 증가에 따라서 이음과 함께 진동발생',u'임펠러부 유체의 불완전화력'),
            createRule(u'모터구동원심압축기 부하증가에 따라 진동증가, 주진동 주파수가 부하에 의해 변동',u'기어부하에 의한 로터, 베어링계의 고유진동수 변화를 간과함, 기어전달 오차의 가진력과 공진'),
            createRule(u'수직펌프 진동이 크다',u'대형 증기터빈의 진동과 펌프 구조계의 공진')]
DB = dict()
for i in range(0,10):
    DB = createRule.addDB(rule[i],DB)
    
MyPrettyPrinter().pprint(inferNov(u'BFP 반구동측의 진동증가',DB))
```

## 6. 충돌 해법

__조건__을 만족하는 두개 이상이 있을 경우의 상황을 가정

__IF__ 조건이 '신호등'이 빨간색이다 일 경우 두 개의 규칙이 동시에 점화될 것, 이러한 규칙의 집합을 __충돌 집합(conflict set)__ 이라 하며 여러 규칙 중 어떤 규칙을 선택할지 결정하는 방법을 __충돌 해법(conflict resolution)__이라고 함

__신호등이 빨간색이라면 어떤 규칙을 수행해야 하는가__

- 규칙의 순서에 따라 결정되는 방법
- 우선 순위를 설정하여 점화하는 방법
- 가장 특수한 규칙을 점화하는 방법
- 데이터베이스에 가장 최근에 입력된 데이터를 사용하는 방법
- ... 대부분의 경우는 위의 방법을 사용하여 거의 만족할 수 있는 해답을 제공한다

__메타지식__

전문가 시스템에 수록된 지식에 관한 지식, 즉 메타지식을 통해 전문가 시스템의 성능을 개선할 수 있음. 메타지식은 지식에 관한 지식(knowledge about knowledge)로 정의 됨
